```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


hastaliklar = ['Akdeniz Ateşi', 'Çölyak', 'Behçet', 'Kistik Fibrozis']
yas_gruplari = ['1-20', '21-40', '41-60']

gorulme_oranlari = [
    [15, 50, 25],   # Akdeniz Ateşi
    [25, 55, 20],   # Çölyak
    [5, 75, 20],    # Behçet
    [60, 30, 10],   # Kistik Fibrozis
]

kadın_oran_yuzde = [40, 70, 40, 50]
erkek_oran_yuzde = [60, 30, 60, 50]

# Kadın ve erkek oranlarını uygun şekilde genişletme
kadın_oranlar = np.array(kadın_oran_yuzde).repeat(len(yas_gruplari))
erkek_oranlar = np.array(erkek_oran_yuzde).repeat(len(yas_gruplari))

# Veriyi hazırlama
data = pd.DataFrame({
    'Hastalık': np.repeat(hastaliklar, len(yas_gruplari)),
    'Yaş Grubu': yas_gruplari * len(hastaliklar),
    'Kadın Oranı': np.array(gorulme_oranlari).flatten() * kadın_oranlar / 100,
    'Erkek Oranı': np.array(gorulme_oranlari).flatten() * erkek_oranlar / 100,
})

# 1. Aykırı Değer Tespiti (İzolasyon Ormanı)
model_if = IsolationForest(contamination=0.1)
data['Aykırı Değer'] = model_if.fit_predict(data[['Kadın Oranı', 'Erkek Oranı']])

# Aykırı değerleri görselleştirme
plt.figure(figsize=(10, 6))
plt.scatter(data['Kadın Oranı'], data['Erkek Oranı'], c=data['Aykırı Değer'], cmap='coolwarm')
plt.xlabel('Kadın Oranı')
plt.ylabel('Erkek Oranı')
plt.title('Aykırı Değer Tespiti (İzolasyon Ormanı)')
plt.colorbar(label='Aykırı Değer Durumu (1: Normal, -1: Aykırı)')
plt.show()

# Aykırı verileri filtreleme
data_normal = data[data['Aykırı Değer'] == 1].copy()

# 2. K-means Kümeleme
kmeans = KMeans(n_clusters=3, random_state=42)
data_normal['Küme'] = kmeans.fit_predict(data_normal[['Kadın Oranı', 'Erkek Oranı']])

# Kümeleme sonuçlarını görselleştirme
plt.figure(figsize=(10, 6))
plt.scatter(data_normal['Kadın Oranı'], data_normal['Erkek Oranı'], c=data_normal['Küme'], cmap='viridis')
plt.xlabel('Kadın Oranı')
plt.ylabel('Erkek Oranı')
plt.title('K-means Kümeleme Sonuçları')
plt.colorbar(label='Küme')
plt.show()

# 3. Karar Ağacı Sınıflandırması
# Yaş Grubu'nu sayısal hale getiriyoruz
yas_grubu_map = {'1-20': 0, '21-40': 1, '41-60': 2}
data_normal['Yaş Grubu Sayısal'] = data_normal['Yaş Grubu'].map(yas_grubu_map)

# Özellikler ve hedef değişkeni
X = data_normal[['Kadın Oranı', 'Erkek Oranı', 'Yaş Grubu Sayısal']]
y = data_normal['Hastalık']

# Eğitim ve test verilerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Karar Ağacı Modeli
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Test verisi üzerinde tahmin yapma
y_pred = clf.predict(X_test)

# Sonuçları yazdırma
accuracy = accuracy_score(y_test, y_pred)
print(f"Karar Ağacı Doğruluk Oranı: {accuracy * 100:.2f}%")

# Karar ağacını görselleştirme
from sklearn.tree import plot_tree

plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=['Kadın Oranı', 'Erkek Oranı', 'Yaş Grubu Sayısal'], class_names=hastaliklar, rounded=True)
plt.title("Karar Ağacı Görselleştirmesi")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Veri seti
data = {
    'Hastalık': ['Akdeniz Ateşi', 'Çölyak', 'Behçet', 'Kistik Fibrozis'],
    'Genetik_Etki': [70, 85, 60, 90]
}

df = pd.DataFrame(data)
print("Veri Seti:")
print(df)

# 1. Isolation Forest ile Aykırı Değer Tespiti
model = IsolationForest(contamination=0.25, random_state=42)
df['Aykırı'] = model.fit_predict(df[['Genetik_Etki']])

# Aykırı değerlerin görünümü
print("\nAykırı Değerler (Isolation Forest):")
print(df[['Hastalık', 'Genetik_Etki', 'Aykırı']])

# 2. K-Means ile Kümeleme
kmeans = KMeans(n_clusters=2, random_state=42)
df['Kume'] = kmeans.fit_predict(df[['Genetik_Etki']])

# Küme merkezlerini yazdırma
print("\nKüme Merkezleri (K-Means):")
print(kmeans.cluster_centers_)

# Küme etiketlerinin veriye eklenmesi
print("\nKüme Etiketleri (K-Means):")
print(df[['Hastalık', 'Genetik_Etki', 'Kume']])

# 3. Karar Ağacı ile Küme Açıklaması
tree = DecisionTreeClassifier(max_depth=2, random_state=42)
tree.fit(df[['Genetik_Etki']], df['Kume'])

# Karar ağacını çizme
plt.figure(figsize=(6, 4))
plot_tree(tree, feature_names=['Genetik_Etki'], class_names=['Küme 0', 'Küme 1'], filled=True)
plt.title("Karar Ağacı ile Küme Açıklaması")
plt.show()

# 4. Genetik Etki Oranı Çizgi Grafiği
plt.figure(figsize=(8, 5))
plt.plot(df['Hastalık'], df['Genetik_Etki'], marker='o', linestyle='-', linewidth=2, color='purple')
for i, oran in enumerate(df['Genetik_Etki']):
    plt.text(df['Hastalık'][i], oran + 1, f"{oran}%", ha='center', fontsize=9)

plt.title('Hastalıklarda Genetik Faktör Etki Oranı', fontsize=13)
plt.xlabel('Hastalıklar')
plt.ylabel('Genetik Etki (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(50, 100)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Veriler
bolgeler = ['Akdeniz', 'İç Anadolu', 'Güneydoğu Anadolu', 'Ege']
etiketler = ['Akdeniz Ateşi', 'Çölyak', 'Behçet', 'Kistik Fibrozis']
oranlar = [30, 25, 25, 20]  # Hastalık oranları
renkler = ['#FF9999', '#99CCFF', '#FFCC99', '#B3FFB3']

# Verileri DataFrame'e dönüştürme
data = pd.DataFrame({
    'Bölge': bolgeler,
    'Hastalık': etiketler,
    'Görülme Oranı': oranlar
})

# Veriyi standardize etme
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['Görülme Oranı']])

# 1. K-means Kümeleme
kmeans = KMeans(n_clusters=2, random_state=42)
data['Küme'] = kmeans.fit_predict(data_scaled)

# 2. Karar Ağacı
X = data[['Görülme Oranı']]  # Özellikler
y = data['Küme']  # Küme etiketleri
tree = DecisionTreeClassifier(max_depth=2, random_state=42)
tree.fit(X, y)

# 3. Isolation Forest (Aykırı Değer Tespiti)
iso_forest = IsolationForest(contamination=0.1, random_state=42)
data['Aykırı Değer'] = iso_forest.fit_predict(data_scaled)
data['Aykırı Değer'] = data['Aykırı Değer'].map({1: 'Normal', -1: 'Aykırı'})

# Veriyi görselleştirme
fig, ax = plt.subplots(figsize=(10, 6))

# K-means Kümeleme Sonuçlarını Görselleştirme
scatter = ax.scatter(data['Bölge'], data['Görülme Oranı'], c=data['Küme'], cmap='viridis', s=100)
ax.set_title("K-means Kümeleme Sonuçları")
ax.set_xlabel("Bölge")
ax.set_ylabel("Görülme Oranı (%)")
plt.xticks(rotation=45)
plt.colorbar(scatter, label="Küme")
plt.tight_layout()

# Karar Ağacını Görselleştirme
fig2, ax2 = plt.subplots(figsize=(10, 6))
plot_tree(tree, filled=True, feature_names=['Görülme Oranı'], class_names=[f'Küme {i}' for i in range(2)], ax=ax2)
ax2.set_title("Karar Ağacı Sonuçları")
plt.tight_layout()

# Aykırı Değerlerin Görselleştirilmesi
fig3, ax3 = plt.subplots(figsize=(10, 6))
colors = {'Normal': '#66CC66', 'Aykırı': '#FF6666'}
ax3.scatter(data['Bölge'], data['Görülme Oranı'], c=data['Aykırı Değer'].map(colors), s=100)
ax3.set_title("Aykırı Değer Tespiti (Isolation Forest)")
ax3.set_xlabel("Bölge")
ax3.set_ylabel("Görülme Oranı (%)")
plt.xticks(rotation=45)
plt.tight_layout()

# Pasta Grafiği - Hastalıkların Görülme Oranı
fig4, ax4 = plt.subplots(figsize=(8, 8))
ax4.pie(data['Görülme Oranı'], labels=data['Hastalık'], autopct='%1.1f%%', colors=renkler, startangle=90, wedgeprops={'edgecolor': 'black'})
ax4.set_title('Hastalıkların Bölgelere Göre Görülme Oranı')

# Grafiklerin gösterilmesi
plt.show()
```
