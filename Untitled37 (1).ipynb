```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


hastaliklar = ['Akdeniz Atei', 'lyak', 'Behet', 'Kistik Fibrozis']
yas_gruplari = ['1-20', '21-40', '41-60']

gorulme_oranlari = [
    [15, 50, 25],   # Akdeniz Atei
    [25, 55, 20],   # lyak
    [5, 75, 20],    # Behet
    [60, 30, 10],   # Kistik Fibrozis
]

kadn_oran_yuzde = [40, 70, 40, 50]
erkek_oran_yuzde = [60, 30, 60, 50]

# Kadn ve erkek oranlarn uygun ekilde geniletme
kadn_oranlar = np.array(kadn_oran_yuzde).repeat(len(yas_gruplari))
erkek_oranlar = np.array(erkek_oran_yuzde).repeat(len(yas_gruplari))

# Veriyi hazrlama
data = pd.DataFrame({
    'Hastalk': np.repeat(hastaliklar, len(yas_gruplari)),
    'Ya Grubu': yas_gruplari * len(hastaliklar),
    'Kadn Oran': np.array(gorulme_oranlari).flatten() * kadn_oranlar / 100,
    'Erkek Oran': np.array(gorulme_oranlari).flatten() * erkek_oranlar / 100,
})

# 1. Aykr Deer Tespiti (zolasyon Orman)
model_if = IsolationForest(contamination=0.1)
data['Aykr Deer'] = model_if.fit_predict(data[['Kadn Oran', 'Erkek Oran']])

# Aykr deerleri grselletirme
plt.figure(figsize=(10, 6))
plt.scatter(data['Kadn Oran'], data['Erkek Oran'], c=data['Aykr Deer'], cmap='coolwarm')
plt.xlabel('Kadn Oran')
plt.ylabel('Erkek Oran')
plt.title('Aykr Deer Tespiti (zolasyon Orman)')
plt.colorbar(label='Aykr Deer Durumu (1: Normal, -1: Aykr)')
plt.show()

# Aykr verileri filtreleme
data_normal = data[data['Aykr Deer'] == 1].copy()

# 2. K-means Kmeleme
kmeans = KMeans(n_clusters=3, random_state=42)
data_normal['Kme'] = kmeans.fit_predict(data_normal[['Kadn Oran', 'Erkek Oran']])

# Kmeleme sonularn grselletirme
plt.figure(figsize=(10, 6))
plt.scatter(data_normal['Kadn Oran'], data_normal['Erkek Oran'], c=data_normal['Kme'], cmap='viridis')
plt.xlabel('Kadn Oran')
plt.ylabel('Erkek Oran')
plt.title('K-means Kmeleme Sonular')
plt.colorbar(label='Kme')
plt.show()

# 3. Karar Aac Snflandrmas
# Ya Grubu'nu saysal hale getiriyoruz
yas_grubu_map = {'1-20': 0, '21-40': 1, '41-60': 2}
data_normal['Ya Grubu Saysal'] = data_normal['Ya Grubu'].map(yas_grubu_map)

# zellikler ve hedef deikeni
X = data_normal[['Kadn Oran', 'Erkek Oran', 'Ya Grubu Saysal']]
y = data_normal['Hastalk']

# Eitim ve test verilerine ayrma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Karar Aac Modeli
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Test verisi zerinde tahmin yapma
y_pred = clf.predict(X_test)

# Sonular yazdrma
accuracy = accuracy_score(y_test, y_pred)
print(f"Karar Aac Doruluk Oran: {accuracy * 100:.2f}%")

# Karar aacn grselletirme
from sklearn.tree import plot_tree

plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=['Kadn Oran', 'Erkek Oran', 'Ya Grubu Saysal'], class_names=hastaliklar, rounded=True)
plt.title("Karar Aac Grselletirmesi")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Veri seti
data = {
    'Hastalk': ['Akdeniz Atei', 'lyak', 'Behet', 'Kistik Fibrozis'],
    'Genetik_Etki': [70, 85, 60, 90]
}

df = pd.DataFrame(data)
print("Veri Seti:")
print(df)

# 1. Isolation Forest ile Aykr Deer Tespiti
model = IsolationForest(contamination=0.25, random_state=42)
df['Aykr'] = model.fit_predict(df[['Genetik_Etki']])

# Aykr deerlerin grnm
print("\nAykr Deerler (Isolation Forest):")
print(df[['Hastalk', 'Genetik_Etki', 'Aykr']])

# 2. K-Means ile Kmeleme
kmeans = KMeans(n_clusters=2, random_state=42)
df['Kume'] = kmeans.fit_predict(df[['Genetik_Etki']])

# Kme merkezlerini yazdrma
print("\nKme Merkezleri (K-Means):")
print(kmeans.cluster_centers_)

# Kme etiketlerinin veriye eklenmesi
print("\nKme Etiketleri (K-Means):")
print(df[['Hastalk', 'Genetik_Etki', 'Kume']])

# 3. Karar Aac ile Kme Aklamas
tree = DecisionTreeClassifier(max_depth=2, random_state=42)
tree.fit(df[['Genetik_Etki']], df['Kume'])

# Karar aacn izme
plt.figure(figsize=(6, 4))
plot_tree(tree, feature_names=['Genetik_Etki'], class_names=['Kme 0', 'Kme 1'], filled=True)
plt.title("Karar Aac ile Kme Aklamas")
plt.show()

# 4. Genetik Etki Oran izgi Grafii
plt.figure(figsize=(8, 5))
plt.plot(df['Hastalk'], df['Genetik_Etki'], marker='o', linestyle='-', linewidth=2, color='purple')
for i, oran in enumerate(df['Genetik_Etki']):
    plt.text(df['Hastalk'][i], oran + 1, f"{oran}%", ha='center', fontsize=9)

plt.title('Hastalklarda Genetik Faktr Etki Oran', fontsize=13)
plt.xlabel('Hastalklar')
plt.ylabel('Genetik Etki (%)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(50, 100)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Veriler
bolgeler = ['Akdeniz', ' Anadolu', 'Gneydou Anadolu', 'Ege']
etiketler = ['Akdeniz Atei', 'lyak', 'Behet', 'Kistik Fibrozis']
oranlar = [30, 25, 25, 20]  # Hastalk oranlar
renkler = ['#FF9999', '#99CCFF', '#FFCC99', '#B3FFB3']

# Verileri DataFrame'e dntrme
data = pd.DataFrame({
    'Blge': bolgeler,
    'Hastalk': etiketler,
    'Grlme Oran': oranlar
})

# Veriyi standardize etme
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['Grlme Oran']])

# 1. K-means Kmeleme
kmeans = KMeans(n_clusters=2, random_state=42)
data['Kme'] = kmeans.fit_predict(data_scaled)

# 2. Karar Aac
X = data[['Grlme Oran']]  # zellikler
y = data['Kme']  # Kme etiketleri
tree = DecisionTreeClassifier(max_depth=2, random_state=42)
tree.fit(X, y)

# 3. Isolation Forest (Aykr Deer Tespiti)
iso_forest = IsolationForest(contamination=0.1, random_state=42)
data['Aykr Deer'] = iso_forest.fit_predict(data_scaled)
data['Aykr Deer'] = data['Aykr Deer'].map({1: 'Normal', -1: 'Aykr'})

# Veriyi grselletirme
fig, ax = plt.subplots(figsize=(10, 6))

# K-means Kmeleme Sonularn Grselletirme
scatter = ax.scatter(data['Blge'], data['Grlme Oran'], c=data['Kme'], cmap='viridis', s=100)
ax.set_title("K-means Kmeleme Sonular")
ax.set_xlabel("Blge")
ax.set_ylabel("Grlme Oran (%)")
plt.xticks(rotation=45)
plt.colorbar(scatter, label="Kme")
plt.tight_layout()

# Karar Aacn Grselletirme
fig2, ax2 = plt.subplots(figsize=(10, 6))
plot_tree(tree, filled=True, feature_names=['Grlme Oran'], class_names=[f'Kme {i}' for i in range(2)], ax=ax2)
ax2.set_title("Karar Aac Sonular")
plt.tight_layout()

# Aykr Deerlerin Grselletirilmesi
fig3, ax3 = plt.subplots(figsize=(10, 6))
colors = {'Normal': '#66CC66', 'Aykr': '#FF6666'}
ax3.scatter(data['Blge'], data['Grlme Oran'], c=data['Aykr Deer'].map(colors), s=100)
ax3.set_title("Aykr Deer Tespiti (Isolation Forest)")
ax3.set_xlabel("Blge")
ax3.set_ylabel("Grlme Oran (%)")
plt.xticks(rotation=45)
plt.tight_layout()

# Pasta Grafii - Hastalklarn Grlme Oran
fig4, ax4 = plt.subplots(figsize=(8, 8))
ax4.pie(data['Grlme Oran'], labels=data['Hastalk'], autopct='%1.1f%%', colors=renkler, startangle=90, wedgeprops={'edgecolor': 'black'})
ax4.set_title('Hastalklarn Blgelere Gre Grlme Oran')

# Grafiklerin gsterilmesi
plt.show()
```
